## 5) Decide on how to address the missing values and do it! (This requires your judgement, so explain your choice).

```python
# drop Outlet Size column
df.drop('Outlet_Size', axis=1, inplace=True)
```

```python
#Creating lookup table for Item weight impute value
df_id_wt = df[['Item_Identifier', 'Item_Weight' ]]
#pd.set_option("display.max_rows", None)
#df_id_wt.sort_values(by=['Item_Identifier'])
```

```python
df_id_wt = df_id_wt.dropna() #removing na
df_id_wt = df_id_wt.drop_duplicates() #removing duplicates
```

```python
df_Id_lookup = df_id_wt
```

```python
df_Id_lookup
```

```python
#filling values for Nan using lookup table to impute Nan's
df['Item_Weight'] = np.where(df['Item_Weight'].isna(),df['Item_Identifier'].map(df_Id_lookup.set_index('Item_Identifier')['Item_Weight']),df['Item_Weight'])
```

```python
df.info()
```

```python
df
```

```python
# drop Nan in Item Weight column which values cannot find in lookup table
df = df.dropna(subset=['Item_Weight'])
```

```python
df.info()
```

```python
df
```

```python
#making sure everything is clean dropna for the last time
df = df.dropna()
```
